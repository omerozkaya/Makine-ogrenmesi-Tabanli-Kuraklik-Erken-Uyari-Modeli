```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.dates as mdates
```

```python
data0=pd.read_csv("05_Nasapower_Climate_Monthly_Mersin_Finalversion.csv")
```

```python
data0.reset_index(drop=True)
```

```python
data0['Date'] = pd.to_datetime(data0['Date'])
```

```python
data0 = data0.sort_values(by='Date')
```

```python
data0.to_csv("04_Nasapower_Climate_Monthly_Mersin_Preprocessed.csv",index=False)
```

```python
data0
```

```python
fig, axes = plt.subplots(ncols=2, nrows=3, figsize=(12, 18))
# 'axes' 2 boyutlu bir dizidir. (axes[0,0], axes[0,1] vb.)
# Sizin kodunuzdaki gibi [0], [1], [2]... diye erişmek için
# bu diziyi 'flatten()' ile düzleştiririz.
ax_list = axes.flatten()

sns.set_style("darkgrid")

# --- GRAFİK 1: Bağıl Nem ---
y_col1 = "2m Relative Humidity (%)"
sns.lineplot(data=data0, x="Date", y=y_col1, ax=ax_list[0])
# plt.xlabel yerine ax_list[0].set_xlabel kullanmalısınız
ax_list[0].set_xlabel("Date")
ax_list[0].set_ylabel(y_col1)
ax_list[0].xaxis.set_major_locator(mdates.YearLocator(base=5))
ax_list[0].xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

# --- GRAFİK 2: Hava Sıcaklığı ---
y_col2 = "2m Air Temperature (C)"
sns.lineplot(data=data0, x="Date", y=y_col2, ax=ax_list[1])
ax_list[1].set_xlabel("Date")
ax_list[1].set_ylabel(y_col2)
ax_list[1].xaxis.set_major_locator(mdates.YearLocator(base=5))
ax_list[1].xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

# --- GRAFİK 3: Rüzgar Hızı ---
y_col3 = "2m Wind Speed (m/s)"
sns.lineplot(data=data0, x="Date", y=y_col3, ax=ax_list[2])
ax_list[2].set_xlabel("Date")
ax_list[2].set_ylabel(y_col3)
ax_list[2].xaxis.set_major_locator(mdates.YearLocator(base=5))
ax_list[2].xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

# --- GRAFİK 4: Güneş Radyasyonu ---
y_col4 = "Global Solar Radiation (MJ/m2/day)"
sns.lineplot(data=data0, x="Date", y=y_col4, ax=ax_list[3])
ax_list[3].set_xlabel("Date")
ax_list[3].set_ylabel(y_col4)
ax_list[3].xaxis.set_major_locator(mdates.YearLocator(base=5))
ax_list[3].xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

# --- GRAFİK 5: YENİ GRAFİK 1 ---
# Buraya 5. grafiğinizin 'y' sütun adını yazın
y_col5 = "Total Precipitation (mm)" 
sns.lineplot(data=data0, x="Date", y=y_col5, ax=ax_list[4])
ax_list[4].set_xlabel("Date")
ax_list[4].set_ylabel(y_col5)
ax_list[4].xaxis.set_major_locator(mdates.YearLocator(base=5))
ax_list[4].xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

# plt.title() SADECE SON GRAFİĞE başlık atar.
# Tüm figür için 'fig.suptitle' 
fig.suptitle("Annual Trend of Climate Parameters (1981–2025) – NASA POWER Data", fontsize=16)
fig.tight_layout()
fig.delaxes(ax_list[5])
# Başlığın (suptitle) grafikle çakışmaması için yer bırak
fig.subplots_adjust(top=0.95) 

# --- 4. Kaydetme ---
# Kaydederken 'fig' nesnesini kullanmak daha iyidir
fig.savefig("January_Trend6.png", dpi=300)

# plt.show()
```
