```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Varsayım: Sütun başlıkları dosyanın 23. satırındadır.
# Pandas'ta bu, 0'dan sayıldığı için header=22'dir (yani 23. satır)

data0 = pd.read_csv(
    "01_Nasapower_Climate_Monthly_Mersin_Raw.csv", 
    sep=',',         # Ayırıcıyı kesin olarak VIRGÜL (,) olarak belirle
    header=13        # Sütun başlıkları dosyanın 23. satırında (indeks 22)
)

# Eğer 23. satır yerine 24. satırdan veriler başlıyorsa,
# bir satır daha atlamanız gerekebilir (skiprows=23 gibi)
```

```python
# Tüm sütunların yeni isimleri, sırasına göre listelenmelidir
yeni_sutun_isimleri = [
    'Parametre', 'Yıl', 'Ocak', 'Şubat', 'Mart', 'Nisan', 'Mayıs', 
    'Haziran', 'Temmuz', 'Ağustos', 'Eylül', 'Ekim', 'Kasım', 'Aralık', 
    'Yıllık Ortalama Toplam'
]

data0.columns = yeni_sutun_isimleri
```

```python
parametre_degistirme_sozlugu = {
    'ALLSKY_SFC_SW_DWN': 'Küresel Güneş Radyasyonu',
    'PRECTOTCORR_SUM': 'Toplam Yağış Miktarı',
    'RH2M': '2m Bağıl Nem',
    'T2M': '2m Hava Sıcaklığı',
    'WS2M': '2m Rüzgâr Hızı'
}

# 'PARAMETER' sütununu al, map() ile değerleri değiştir ve yeni değerleri geri ata.
data0["Parametre"] = data0["Parametre"].map(parametre_degistirme_sozlugu)
```

```python
data0["Parametre"].unique()
```

```python
data0.head()
```

```python
aylik_sutunlar = ['Ocak', 'Şubat', 'Mart', 'Nisan', 'Mayıs', 'Haziran', 
                    'Temmuz', 'Ağustos', 'Eylül', 'Ekim', 'Kasım', 'Aralık','Yıllık Ortalama Toplam']

for col in aylik_sutunlar:
    data0[col] = data0[col].replace(-999.00, data0[col].median())
```

```python
data1 = pd.read_csv("02_Nasapower_Climate_Monthly_Mersin_Preprocessed.csv")
```

```python
aylik_sutunlar = ["Ocak", "Şubat", "Mart", "Nisan", "Mayıs", "Haziran", 
                  "Temmuz", "Ağustos", "Eylül", "Ekim", "Kasım", "Aralık"]

data1_melted = data1.melt(
    # Kimlik Değişkenleri (Değişmeyecek olanlar)
    id_vars=["Parametre", "Yıl", "Yıllık Ortalama Toplam"], 
    
    # Eritilecek Değişkenler (Aylık Sütunlar)
    value_vars=aylik_sutunlar,
    
    # Eriyen Sütun Adlarının Toplanacağı Yeni Sütun
    var_name="Ay_Adı",
    
    # Parametre Değerlerinin Toplanacağı Yeni Sütun
    value_name="Değer" 
)

# Sonuçta data1_melted DataFrame'inizde 5 sütun olacaktır:
# [Parametre, Yıl, Yıllık Ortalama Toplam, Ay_Adı, Parametre_Değeri]
```

```python
data1_melted
```

```python
# Ayların metin karşılıkları ve sayısal karşılıkları
aylik_sutunlar = ["Ocak", "Şubat", "Mart", "Nisan", "Mayıs", "Haziran", 
                  "Temmuz", "Ağustos", "Eylül", "Ekim", "Kasım", "Aralık"]

# Ay adı - Ay sayısı eşleştirmesi
ay_map = {ay: i + 1 for i, ay in enumerate(aylik_sutunlar)}

# Yeni sütun oluşturma: Ay adlarını sayısal değere çevirir
data1_melted['Ay_Sayı'] = data1_melted['Ay_Adı'].map(ay_map)
```

```python
data1_melted
```

```python
# data1_melted DataFrame'indeki Yıl ve Ay_Sayı sütunlarını kullanıyoruz
data1_melted['Tarih'] = (
    data1_melted['Yıl'].astype(str) + '-' + 
    data1_melted['Ay_Sayı'].astype(str).str.zfill(2) # Ay sayısını iki haneli yapar (örn: '1' yerine '01')
)
```

```python
data1_melted
```

```python
data1_melted.drop(["Ay_Adı","Ay_Sayı"],axis=1)
```

```python
data1_melted.to_csv('03_Nasapower_Climate_Monthly_Mersin_Preprocessed.csv', 
    index=False,        # Satır indeksini (0, 1, 2...) dosyaya kaydetmez. Bu genellikle istenir.
    encoding='utf-8'    # Türkçe karakterler için genellikle en iyi seçimdir.
                   )
```

```python
dataset2=pd.read_csv("03_Nasapower_Climate_Monthly_Mersin_Preprocessed.csv")
```

```python
dataset2["Parametre"].unique().sum()
```

```python
dataset2
```

```python
dataset3 = dataset2.pivot_table(
    index="Tarih",      # Yeni Tablonun Satırları
    columns='Parametre',           # Yeni Tablonun Sütun Başlıkları
    values="Değer"                # Hücrelere Yerleşecek Veriler
)
dataset3.columns.name = None
```

```python
dataset3
```

```python
# Sütun adlarını ve birimlerini tanımlayan sözlük
yeni_sutun_adlari = {
    '2m Bağıl Nem': '2m Bağıl Nem (%)',
    '2m Hava Sıcaklığı': '2m Hava Sıcaklığı (C)', # °C sembolü yerine C kullanmak dosya uyumluluğu için daha iyidir
    '2m Rüzgâr Hızı': '2m Rüzgâr Hızı (m/s)',
    'Küresel Güneş Radyasyonu': 'Küresel Güneş Radyasyonu (MJ/m2/gün)', # MJ/m² ve /gün eklenmiştir
    'Toplam Yağış Miktarı': 'Toplam Yağış Miktarı (mm)' # Aylık toplam olduğunu varsayıyoruz
}

# DataFrame'deki sütun adlarını yeniden adlandırma
dataset3 = dataset3.rename(columns=yeni_sutun_adlari)
```

```python
dataset3.to_csv("05_Nasapower_Climate_Monthly_Mersin_Finalversion.csv")
```

```python
dataset3corr=dataset3.corr()
```

```python
sns.heatmap(
    dataset3corr,
    annot=True,          # Korelasyon değerlerini hücrelerin üzerine yaz
    cmap='coolwarm',     # Renk paleti
    cbar_kws={'label': 'Korelasyon Katsayısı'}
)
plt.title("Korelasyon Isı Haritası (1981-2025) -Düzenlenmiş NASA POWER Verisi")
plt.savefig("Korelasyon_Heatmap4.png")
```
