```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset0 = pd.read_csv("09_Nasapower_Climate_Monthly_Mersin_FinalVersion.csv",index_col=0)
```

```python
dataset0
```

```python
dataset0 = dataset0.reset_index()
```

```python
dataset0.head()
```

```python
dataset0['Date'] = pd.to_datetime(dataset0['Date'], errors='coerce')
dataset0.set_index('Date', inplace=True)
```

```python
dataset1 = dataset0.resample('YE').mean(numeric_only=True)
```

```python
#dataset2=dataset1.loc[:,"2m Hava Sıcaklığı(°C)"]
```

Series (Dizi): dataset1.loc[:, "SütunAdı"] derseniz, Pandas size tek boyutlu bir Series verir. Series'in içinde "sütun ismi" aranmaz, çünkü kendisi zaten o sütundur.

```python
dataset2=dataset1.loc[:,["2m Air Temperature (°C)"]]
```

```python
dataset2
```

```python
plt.figure(figsize=(12, 6))
plt.plot(dataset2, color='#FFD700', linewidth=1.5)
plt.title("Inter-annual Variability of Monthly Mean Temperature", fontsize=14, fontweight='bold', fontname='Arial')
plt.xlabel("Date", fontsize=12, fontname='Arial')
plt.ylabel("Temperature ($^\circ$C)", fontsize=12, fontname='Arial')
plt.grid(True, which='both', linestyle='--', linewidth=0.5, alpha=0.7) # İnce, şık bir ızgara
plt.xticks(rotation=45) # Tarihlerin okunabilirliği
plt.savefig("Interannual_Variability.png")
plt.show()
```

```python
plt.figure(figsize=(18, 6))
plt.bar(dataset0.index, dataset0["Total Precipitation (mm)"], 
        color='blue', alpha=0.5, label='Precipitation', width=25)
plt.plot(dataset0.index, dataset0["Total Precipitation (mm)"], 
         color='red', linewidth=2, label='Trend')
plt.title("Precipitation Volume", fontsize=14)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Total Precipitation (mm)", fontsize=12)
plt.legend(loc='upper right')
plt.xticks(rotation=75)
plt.tight_layout()
plt.show()
```
